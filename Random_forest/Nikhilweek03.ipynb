```python
X_train = df_train.iloc [:, :50]
y_train = df_train.drop(df_train.iloc [:, :50],axis = 1)
X_test = df_test.iloc [:, :50]
y_test = df_test.drop(df_test.iloc [:, :50],axis = 1)
```

# BUILDING RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestRegressor
def predict_rf(d1,d2):
    X_train = d1.iloc [:, :50]
    X_test = d2.iloc [:, :50]
    n_list = [10,100,500,1000]
    y_list = ['quantile_1', 'quantile_5', 'quantile_9','avg','var']
    out = ''

    for l in y_list:
        y_train = df_train[l]
        y_test = df_test[l]
        for n in n_list:
            out += 'Current trees: ' + str(n) + '\tCurrent target feature: ' + str(l) +'\n'
            random_forest = RandomForestRegressor(n_estimators= n)
            random_forest.fit(X_train, y_train)
            y_pred = random_forest.predict(X_test)
            out+= 'SCORE ' + str(random_forest.score(X_test,y_test))+'\n'
    return out
```

```python
output  = predict_rf(df_train,df_test)
```
